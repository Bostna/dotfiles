```python

# Map each TACO label to a Shibuya disposal type
shibuya_map = {
    "Pop tab": "Recyclable: Cans", "Single-use carrier bag": "Recyclable: Plastics", "Tissues": "Burnable",
    "Crisp packet": "Recyclable: Plastics", "Garbage bag": "Recyclable: Plastics", "Rope & strings": "Burnable",
    "Foam cup": "Burnable", "Aerosol": "Recyclable: Spray cans (empty)", "Spread tub": "Recyclable: Plastics",
    "Shoe": "Burnable", "Squeezable tube": "Recyclable: Plastics", "Aluminium blister pack": "Non-burnable",
    "Six pack rings": "Recyclable: Plastics", "Toilet tube": "Recyclable: Paper", "Tupperware": "Recyclable: Plastics",
    "Polypropylene bag": "Recyclable: Plastics", "Pizza box": "Recyclable: Cardboard",
    "Carded blister pack": "Recyclable: Plastics", "Glass jar": "Recyclable: Bottles (glass)"
}
```

```python
# define map function
def map_shibuya_type(name: str) -> str:
    n = str(name)
    if n in shibuya_map: return shibuya_map[n]
    s = n.lower()
    if "pet" in s and "bottle" in s: return "Recyclable: PET bottles"
    if "glass" in s and "bottle" in s: return "Recyclable: Bottles (glass)"
    if "can" in s and not ("spray" in s or "paint" in s): return "Recyclable: Cans"
    if any(k in s for k in ["paper", "newspaper", "magazine", "straw"]): return "Recyclable: Paper"
    if any(k in s for k in ["cardboard", "carton", "corrugated"]): return "Recyclable: Cardboard"
    if "plastic bottle" in s: return "Recyclable: PET bottles"
    if any(k in s for k in ["plastic", "styrofoam", "polystyrene", "wrapper", "packaging", "polypropylene", "lid", "utensil", "container", "film", "cap", "bag"]): return "Recyclable: Plastics"
    if "spray" in s or "aerosol" in s: return "Recyclable: Spray cans (empty)"
    if any(k in s for k in ["battery", "lighter", "aluminium foil", "aluminum foil", "foil", "ceramic", "porcelain", "broken glass", "glass cup", "metal"]): return "Non-burnable"
    if any(k in s for k in ["cigarette", "butt", "rubber", "leather", "clothing", "cloth", "diaper", "sanitary", "food"]): return "Burnable"
    return "Review"
```

```python

def to_main_type(shibuya: str) -> str:
    if pd.isna(shibuya): return "Review"
    s = str(shibuya).strip().lower()
    if s.startswith("recyclable"): return "Recycle"
    if s == "burnable": return "Burnable"
    if s.replace("-", " ") == "non burnable": return "Non-burnable"
    return "Review"
```

```python
# --- Apply the mapping logic ---
counts = df["category_name"].value_counts().rename_axis("category_name").reset_index(name="count")
counts["shibuya_type"] = counts["category_name"].apply(map_shibuya_type)
counts["main_type"] = counts["shibuya_type"].apply(to_main_type)

# Add main_type back to the original dataframe
df = df.merge(counts[["category_name", "main_type"]], on="category_name", how="left")
```

```python
# --- Normalize bounding box coordinates ---
df["x_norm"]  = df["x"] / df["width_img"]
df["y_norm"]  = df["y"] / df["height_img"]
df["w_norm"]  = df["width_bb"] / df["width_img"]
df["h_norm"]  = df["height_bb"] / df["height_img"]
df["xc_norm"] = (df["x"] + df["width_bb"]/2) / df["width_img"]
df["yc_norm"] = (df["y"] + df["height_bb"]/2) / df["height_img"]
```

```python

# --- Visualize some samples ---
print("Displaying sample images with bounding boxes...")
f, ax = plt.subplots(1, 5, figsize=(20, 4))
for i in range(5):
    img_bgr = cv2.imread(df.loc[i, "file_path"])
    if img_bgr is None:
        print(f"Warning: Could not read image at {df.loc[i, 'file_path']}")
        continue # Skip to the next image if read fails
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    H, W = img_rgb.shape[:2]

    xc, yc, wn, hn = df.loc[i, ["xc_norm", "yc_norm", "w_norm", "h_norm"]]
    w, h = int(wn * W), int(hn * H)
    x, y = int(xc * W - w / 2), int(yc * H - h / 2)

    label = f"{df.loc[i, 'category_name']} | {df.loc[i, 'main_type']}"
    cv2.rectangle(img_rgb, (x, y), (x + w, y + h), (255, 0, 0), 3)
    cv2.putText(img_rgb, df.loc[i, 'main_type'], (x, max(0, y - 10)), cv2.FONT_HERSHEY_SIMPLEX,3, (255, 0, 0), 2)
    ax[i].imshow(img_rgb)
    ax[i].axis('off')
plt.tight_layout()
plt.show()
```

# ===============================================================
# 4. TENSORFLOW DATASET PREPARATION
# ===============================================================

```python
# --- Prepare data for TensorFlow ---
df["class_id"] = pd.factorize(df["category_id"])[0].astype(np.int32)
num_classes = df["class_id"].nunique()

X_paths = df["file_path"].astype(str).values
Y_box_norm = df[["xc_norm", "yc_norm", "w_norm", "h_norm"]].astype(np.float32).values
Y_cls = df["class_id"].astype(np.int32).values
```

```python

# Network input size
IMG_WIDTH, IMG_HEIGHT = 512, 512

def read_image_bbox(path, bbox_norm, class_id):
    path = path.decode()
    img = cv2.imread(path)
    if img is None:
        img = np.zeros((IMG_HEIGHT, IMG_WIDTH, 3), np.uint8) # Dummy image
    else:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))

    xc, yc, wn, hn = bbox_norm
    x1, y1 = float(xc - wn/2), float(yc - hn/2)
    x2, y2 = float(xc + wn/2), float(yc + hn/2)
    norm_xyxy = np.array([x1, y1, x2, y2], dtype=np.float32)

    return (img.astype(np.float32) / 255.0, norm_xyxy, np.int32(class_id))
```

```python
def parse(path, bbox_norm, class_id):
    img, bbox, cid = tf.numpy_function(
        read_image_bbox, [path, bbox_norm, class_id], [tf.float32, tf.float32, tf.int32]
    )
    img.set_shape((IMG_HEIGHT, IMG_WIDTH, 3))
    bbox.set_shape((4,))
    cid.set_shape(())
    onehot = tf.one_hot(cid, num_classes)
    return img, (bbox, onehot)
```

```python

def tf_dataset(paths, boxes, labels, batch=16, shuffle=True): # Increased batch size for GPU
    ds = tf.data.Dataset.from_tensor_slices((paths, boxes, labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(paths))
    ds = ds.map(parse, num_parallel_calls=tf.data.AUTOTUNE)
    ds = ds.batch(batch).prefetch(tf.data.AUTOTUNE)
    return ds
```

```python

# --- Create Train/Validation/Test Splits ---
n = len(X_paths)
n_train = int(0.8 * n)
n_valid = int(0.1 * n)

train_ds = tf_dataset(X_paths[:n_train], Y_box_norm[:n_train], Y_cls[:n_train])
valid_ds = tf_dataset(X_paths[n_train:n_train+n_valid], Y_box_norm[n_train:n_train+n_valid], Y_cls[n_train:n_train+n_valid], shuffle=False)
test_ds  = tf_dataset(X_paths[n_train+n_valid:], Y_box_norm[n_train+n_valid:], Y_cls[n_train+n_valid:], shuffle=False)

print(f"Dataset created: {len(train_ds)} training batches, {len(valid_ds)} validation batches.")
```

```python
from tensorflow.keras.models import Model # Import Model for functional API
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# model 1 (using Functional API for multiple outputs)
input_tensor = Input(shape=(IMG_HEIGHT, IMG_WIDTH, 3))

x = Conv2D(filters = 32, kernel_size = (3,3), activation="relu", padding="same")(input_tensor)
x = MaxPooling2D(pool_size=(2, 2), padding="same")(x)

x = Conv2D(filters = 32, kernel_size = (3,3), activation="relu", padding="same")(x)
x = MaxPooling2D(pool_size=(2, 2), padding="same")(x)

x = Conv2D(filters = 64, kernel_size = (3,3), activation="relu", padding="same")(x)
x = MaxPooling2D(pool_size=(2, 2), padding="same")(x)

x = Conv2D(filters = 128, kernel_size = (3,3), activation="relu", padding="same")(x)
x = MaxPooling2D(pool_size=(2, 2), padding="same")(x)

# Flatten the output for the dense layers
flatten = Flatten()(x)

# Bounding box regression head
bbox_head = Dense(64, activation="relu")(flatten)
bbox_head = Dense(4, name="bbox_output")(bbox_head) # Output layer for bounding box (x1, y1, x2, y2)

# Classification head
class_head = Dense(64, activation="relu")(flatten)
class_head = Dropout(0.5)(class_head)
class_head = Dense(num_classes, activation="softmax", name="class_output")(class_head) # Output layer for class probabilities

# Create the model with two outputs
model = Model(inputs=input_tensor, outputs=[bbox_head, class_head])
```

```python
model.summary()
```

```python
adam = optimizers.Adam(learning_rate = 0.001)
model.compile(optimizer=adam,
              loss={'bbox_output': 'mse', 'class_output': 'categorical_crossentropy'}, # Specify loss for each output
              metrics={'bbox_output': 'mse', 'class_output': 'accuracy'}) # Specify metrics for each output
```

```python
from tensorflow.keras import callbacks

# Callbacks
MODEL = "model_1.keras"

modelCheckpoint = callbacks.ModelCheckpoint(MODEL,
                                            monitor="val_loss",
                                            verbose=0,
                                            save_best_only=True)

LRreducer = callbacks.ReduceLROnPlateau(monitor="val_loss",
                                        factor=0.1,
                                        patience=3,
                                        verbose=1,
                                        min_lr=0)

EarlyStopper = callbacks.EarlyStopping(monitor='val_loss',
                                       patience=10,
                                       verbose=0,
                                       restore_best_weights=True)
```

```python
#  model training
%%time
history = model.fit(
        train_ds,
        epochs=30,
        validation_data=valid_ds,
        callbacks=[modelCheckpoint, LRreducer, EarlyStopper]
        )
```
